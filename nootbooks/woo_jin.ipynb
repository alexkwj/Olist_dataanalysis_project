## Identify orders from repeat customers

### Subtask:
Filter the `orders_dataset` to identify orders placed by repeat customers. Repeat customers are defined as those with more than one entry in the `customer_unique_id` column of the `customers` DataFrame.


**Reasoning**:
I need to identify repeat customers and then filter the orders dataset based on these repeat customers. I will use the `value_counts()` method to find customers with more than one order and then filter the `customers` DataFrame to get their `customer_id`s. Finally, I will use these `customer_id`s to filter the `orders_dataset`.



```python
# Step 1: Identify repeat customers
customer_counts = customers['customer_unique_id'].value_counts()
repeat_customer_unique_ids = customer_counts[customer_counts > 1].index
repeat_customer_ids = customers[customers['customer_unique_id'].isin(repeat_customer_unique_ids)]['customer_id']

# Step 2: Filter orders_dataset for orders from repeat customers
repeat_customer_orders = orders_dataset[orders_dataset['customer_id'].isin(repeat_customer_ids)]

# Step 3: Display the first few rows and the shape of repeat_customer_orders
display(repeat_customer_orders.head())
print(repeat_customer_orders.shape)
```

```python
repeat_customer_products = pd.merge(repeat_customer_orders, order_items[['order_id', 'product_id']], on='order_id', how='inner')
repeat_customer_products = repeat_customer_products[['order_id', 'product_id']]
display(repeat_customer_products.head())
print(repeat_customer_products.shape)
```

```python
# Convert timestamp columns to datetime objects
orders_dataset['order_purchase_timestamp'] = pd.to_datetime(orders_dataset['order_purchase_timestamp'])
orders_dataset['order_estimated_delivery_date'] = pd.to_datetime(orders_dataset['order_estimated_delivery_date'])
orders_dataset['order_delivered_customer_date'] = pd.to_datetime(orders_dataset['order_delivered_customer_date'])

# Calculate estimated delivery time
orders_dataset['estimated_delivery_time'] = (orders_dataset['order_estimated_delivery_date'] - orders_dataset['order_purchase_timestamp']).dt.days

# Calculate actual delivery time
orders_dataset['actual_delivery_time'] = (orders_dataset['order_delivered_customer_date'] - orders_dataset['order_purchase_timestamp']).dt.days

# Keep only successfully delivered orders
orders_dataset = orders_dataset[orders_dataset['actual_delivery_time'].notnull()].copy()

# Display the first few rows and the shape of the updated DataFrame
display(orders_dataset.head())
print(orders_dataset.shape)
```

## 1회 구매자 식별

### Subtask:
1회 구매자 식별


**Reasoning**:
Calculate the count of each unique customer_unique_id, identify those with a count of 1, and store these unique IDs in a list.



```python
customer_counts = customers['customer_unique_id'].value_counts()
single_occurrence_unique_ids = customer_counts[customer_counts == 1].index.tolist()
print(f"Number of unique customers with only one order: {len(single_occurrence_unique_ids)}")
```

## 재구매자 식별

### Subtask:
`customers` 데이터프레임을 사용하여 재구매자의 `customer_unique_id`를 식별합니다.


**Reasoning**:
Calculate the value counts of `customer_unique_id` to identify repeat customers, select the index where the count is greater than 1, convert to a list, and print the count.



```python
# Calculate the value counts of `customer_unique_id`
customer_counts = customers['customer_unique_id'].value_counts()

# Select the index where the count is greater than 1 to identify repeat customers
repeat_customer_unique_ids = customer_counts[customer_counts > 1].index

# Convert the index to a list
repeat_customer_unique_ids_list = repeat_customer_unique_ids.tolist()

# Print the count of unique repeat customer IDs
print(f"Number of unique repeat customer_unique_ids: {len(repeat_customer_unique_ids_list)}")
```

## 주문 데이터 분류

### Subtask:
`orders_dataset`을 1회 구매자의 주문과 재구매자의 주문으로 나눕니다.


**Reasoning**:
Filter the orders_dataset based on whether the customer_unique_id is in the single_occurrence_unique_ids list or the repeat_customer_unique_ids_list and display the head and shape of the resulting dataframes.



```python
# Filter orders_dataset for orders from one-time customers
one_time_customer_orders = orders_dataset[orders_dataset['customer_id'].isin(customers[customers['customer_unique_id'].isin(single_occurrence_unique_ids)]['customer_id'])]

# Filter orders_dataset for orders from repeat customers
repeat_customer_orders = orders_dataset[orders_dataset['customer_id'].isin(customers[customers['customer_unique_id'].isin(repeat_customer_unique_ids_list)]['customer_id'])]

# Display the first few rows and the shape of one_time_customer_orders
print("One-time customer orders:")
display(one_time_customer_orders.head())
print(one_time_customer_orders.shape)

# Display the first few rows and the shape of repeat_customer_orders
print("\nRepeat customer orders:")
display(repeat_customer_orders.head())
print(repeat_customer_orders.shape)
```

## 배송 지연율 계산

### Subtask:
각 그룹(1회 구매자, 재구매자)에 대해 예상 배송 시간보다 실제 배송 시간이 더 긴 주문의 비율을 계산합니다.


**Reasoning**:
Calculate the number of delayed orders and the total number of orders for both one-time and repeat customers, then calculate and print the delay rate for each group as per the instructions.



```python
# Step 1: Calculate the number of delayed orders for one-time customers
delayed_one_time_orders_count = one_time_customer_orders[
    one_time_customer_orders['actual_delivery_time'] > one_time_customer_orders['estimated_delivery_time']
].shape[0]

# Step 2: Calculate the total number of orders for one-time customers
total_one_time_orders_count = one_time_customer_orders.shape[0]

# Step 3: Calculate the delivery delay rate for one-time customers
one_time_delay_rate = (delayed_one_time_orders_count / total_one_time_orders_count) * 100 if total_one_time_orders_count > 0 else 0

# Step 4: Calculate the number of delayed orders for repeat customers
delayed_repeat_orders_count = repeat_customer_orders[
    repeat_customer_orders['actual_delivery_time'] > repeat_customer_orders['estimated_delivery_time']
].shape[0]

# Step 5: Calculate the total number of orders for repeat customers
total_repeat_orders_count = repeat_customer_orders.shape[0]

# Step 6: Calculate the delivery delay rate for repeat customers
repeat_delay_rate = (delayed_repeat_orders_count / total_repeat_orders_count) * 100 if total_repeat_orders_count > 0 else 0

# Step 7: Print the calculated delay rates
print(f"1회 구매자의 배송 지연율: {one_time_delay_rate:.2f}%")
print(f"재구매자의 배송 지연율: {repeat_delay_rate:.2f}%")
```

## 데이터 시각화

### Subtask:
계산된 배송 지연율을 사용하여 막대 그래프를 생성합니다.


**Reasoning**:
Generate a bar chart to visualize the calculated delivery delay rates for one-time and repeat customers.



```python
# Create data for the bar chart
labels = ['1회 구매자', '재구매자']
sizes = [one_time_delay_rate, repeat_delay_rate]
colors = ['#FFA500','#32CD32'] # Orange and Green

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(labels, sizes, color=colors)
plt.title('1회 구매자 vs 재구매자의 배송 지연율', fontsize=16) # Increased title font size
plt.ylabel('배송 지연율 (%)', fontsize=14) # Increased ylabel font size
plt.xticks(fontsize=12) # Increased xticks font size
plt.yticks(fontsize=12) # Increased yticks font size
plt.show()
```
