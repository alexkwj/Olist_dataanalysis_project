## 가격대별 분석

### Subtask:
`merged_df`의 'price' 컬럼을 사용하여 상품 가격 분포를 파악하고, 특정 가격대별 주문 건수나 총 판매 금액을 분석합니다. 시각화를 통해 인사이트를 얻습니다.


**Reasoning**:
Calculate descriptive statistics for the 'price' column in the `merged_df` to understand the price distribution and then plot a histogram to visualize it.



```python
# 'price' 열의 기술 통계량 계산
price_stats = merged_df['price'].describe()
print("Price Column Descriptive Statistics:")
display(price_stats)

# 'price' 열의 히스토그램 그리기
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['price'], bins=50, kde=True)
plt.title('상품 가격 분포')
plt.xlabel('가격')
plt.ylabel('주문 건수')
plt.show()
```

**Reasoning**:
Define price categories based on the descriptive statistics and then calculate and visualize the order count and total sales amount for each category.



```python
# 특정 가격대 정의 (예: 25%, 50%, 75% 분위수 사용)
low_price_threshold = merged_df['price'].quantile(0.25)
mid_price_threshold = merged_df['price'].quantile(0.50)
high_price_threshold = merged_df['price'].quantile(0.75)

# 가격대별 카테고리 할당
def categorize_price(price):
    if price <= low_price_threshold:
        return 'Low'
    elif price <= mid_price_threshold:
        return 'Medium-Low'
    elif price <= high_price_threshold:
        return 'Medium-High'
    else:
        return 'High'

merged_df['price_category'] = merged_df['price'].apply(categorize_price)

# 가격대별 주문 건수 계산
price_category_counts = merged_df['price_category'].value_counts().reset_index()
price_category_counts.columns = ['price_category', 'order_count']

# 가격대별 총 판매 금액 계산
price_category_values = merged_df.groupby('price_category')['payment_value'].sum().reset_index()
price_category_values.columns = ['price_category', 'total_sales_value']

# 결과 출력
print("\n가격대별 주문 건수:")
display(price_category_counts)

print("\n가격대별 총 판매 금액:")
display(price_category_values)

# 가격대별 주문 건수 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=price_category_counts, x='price_category', y='order_count', order=['Low', 'Medium-Low', 'Medium-High', 'High'])
plt.title('가격대별 주문 건수')
plt.xlabel('가격대')
plt.ylabel('주문 건수')
plt.show()

# 가격대별 총 판매 금액 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=price_category_values, x='price_category', y='total_sales_value', order=['Low', 'Medium-Low', 'Medium-High', 'High'])
plt.title('가격대별 총 판매 금액')
plt.xlabel('가격대')
plt.ylabel('총 판매 금액')
plt.show()
```

**Reasoning**:
Explore the relationship between product category and freight value to see if certain categories have significantly different shipping costs, addressing the fourth instruction.



```python
# merged_df와 df6(products_dataset)를 'product_id'를 기준으로 병합하여 상품 카테고리 정보 추가
merged_df_with_category = pd.merge(merged_df, df6[['product_id', 'product_category_name']], on='product_id', how='left')

# 상품 카테고리별 평균 배송비 계산
avg_freight_by_category = merged_df_with_category.groupby('product_category_name')['freight_value'].mean().reset_index()
avg_freight_by_category.columns = ['product_category_name', 'average_freight_value']

# 평균 배송비가 높은 상위 10개 카테고리 선택
top_freight_categories = avg_freight_by_category.nlargest(10, 'average_freight_value')

# 결과 출력
print("\n평균 배송비 상위 10개 상품 카테고리:")
display(top_freight_categories)

# 평균 배송비 상위 10개 상품 카테고리 시각화
plt.figure(figsize=(12, 7))
sns.barplot(data=top_freight_categories, x='average_freight_value', y='product_category_name')
plt.title('평균 배송비 상위 10개 상품 카테고리')
plt.xlabel('평균 배송비')
plt.ylabel('상품 카테고리')
plt.tight_layout()
plt.show()
```

Define price categories based on the descriptive statistics and then calculate and visualize the order count and total sales amount for each category.

```python
# 특정 가격대 정의 (예: 25%, 50%, 75% 분위수 사용)
low_price_threshold = merged_df['price'].quantile(0.25)
mid_price_threshold = merged_df['price'].quantile(0.50)
high_price_threshold = merged_df['price'].quantile(0.75)

# 가격대별 카테고리 할당
def categorize_price(price):
    if price <= low_price_threshold:
        return 'Low'
    elif price <= mid_price_threshold:
        return 'Medium-Low'
    elif price <= high_price_threshold:
        return 'Medium-High'
    else:
        return 'High'

merged_df['price_category'] = merged_df['price'].apply(categorize_price)

# 가격대별 주문 건수 계산
price_category_counts = merged_df['price_category'].value_counts().reset_index()
price_category_counts.columns = ['price_category', 'order_count']

# 가격대별 총 판매 금액 계산
price_category_values = merged_df.groupby('price_category')['payment_value'].sum().reset_index()
price_category_values.columns = ['price_category', 'total_sales_value']

# 결과 출력
print("\n가격대별 주문 건수:")
display(price_category_counts)

print("\n가격대별 총 판매 금액:")
display(price_category_values)

# 가격대별 주문 건수 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=price_category_counts, x='price_category', y='order_count', order=['Low', 'Medium-Low', 'Medium-High', 'High'])
plt.title('가격대별 주문 건수')
plt.xlabel('가격대')
plt.ylabel('주문 건수')
plt.show()

# 가격대별 총 판매 금액 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=price_category_values, x='price_category', y='total_sales_value', order=['Low', 'Medium-Low', 'Medium-High', 'High'])
plt.title('가격대별 총 판매 금액')
plt.xlabel('가격대')
plt.ylabel('총 판매 금액')
plt.show()
```

## 재구매율과 배송 지연 관계 분석

### Subtask:
배송 지연이 고객의 재구매율에 미치는 영향을 분석합니다. 배송 지연이 발생한 고객 그룹과 그렇지 않은 고객 그룹의 재구매율을 비교하고 시각화합니다. (재구매율 분석을 위해 고객별 주문 횟수를 계산하고, 첫 주문 시 배송 지연 경험이 재구매로 이어지는지 등을 분석할 수 있습니다.)


**Reasoning**:
고객별 고유 주문 횟수를 계산하고 재구매 고객 여부를 판별하여 `merged_df`에 병합한 후, 첫 주문 시 배송 지연 여부에 따른 재구매율을 계산하고 시각화합니다.



```python
# 1. 고객별 고유 주문 횟수 계산
customer_order_counts = merged_df.groupby('customer_unique_id')['order_id'].nunique().reset_index()
customer_order_counts.columns = ['customer_unique_id', 'order_count']

# 2. 'is_repeat_customer' 컬럼 추가
customer_order_counts['is_repeat_customer'] = customer_order_counts['order_count'] > 1

# 결과 출력 (상위 10개 고객)
print("고객별 주문 횟수 및 재구매 여부 (상위 10개 고객):")
display(customer_order_counts.head(10))

# 3. merged_df에 고객별 주문 횟수 및 재구매 여부 병합
# merged_df에는 한 주문의 여러 아이템이 포함될 수 있으므로, order_id 기준으로 중복 제거 후 병합
merged_orders = merged_df.drop_duplicates(subset=['order_id', 'customer_unique_id'])
merged_orders = pd.merge(merged_orders, customer_order_counts, on='customer_unique_id', how='left')

# 고객의 첫 주문 정보 찾기
first_orders = merged_orders.loc[merged_orders.groupby('customer_unique_id')['order_purchase_timestamp'].idxmin()]

# 4. 첫 주문 시 배송 지연 여부에 따른 재구매율 계산
# 배송 지연 기준 (예: delivery_delay_days > 0)
first_orders['had_delay_on_first_order'] = first_orders['delivery_delay_days'] > 0

# 첫 주문 지연 그룹별 재구매 고객 수 계산
repeat_customer_by_delay = first_orders.groupby('had_delay_on_first_order')['is_repeat_customer'].sum().reset_index()
repeat_customer_by_delay.columns = ['had_delay_on_first_order', 'repeat_customer_count']

# 첫 주문 지연 그룹별 총 고객 수 계산
total_customer_by_delay = first_orders.groupby('had_delay_on_first_order').size().reset_index(name='total_customer_count')

# 재구매율 계산
retention_rate_by_delay = pd.merge(repeat_customer_by_delay, total_customer_by_delay, on='had_delay_on_first_order')
retention_rate_by_delay['retention_rate'] = (retention_rate_by_delay['repeat_customer_count'] / retention_rate_by_delay['total_customer_count']) * 100

# 결과 출력
print("\n첫 주문 시 배송 지연 여부에 따른 재구매율:")
display(retention_rate_by_delay)

# 5. 재구매율 시각화
plt.figure(figsize=(8, 6))
sns.barplot(data=retention_rate_by_delay, x='had_delay_on_first_order', y='retention_rate')
plt.title('첫 주문 시 배송 지연 여부에 따른 재구매율')
plt.xlabel('첫 주문 시 배송 지연 여부 (True: 지연, False: 지연 없음)')
plt.ylabel('재구매율 (%)')
plt.xticks([0, 1], ['지연 없음', '지연'])
plt.ylim(0, retention_rate_by_delay['retention_rate'].max() * 1.1) # Y축 범위 조정
plt.show()
```

```python
# '실제 배송 소요 시간' 컬럼 생성
# 'order_delivered_customer_date'와 'order_purchase_timestamp'의 차이 계산
# 실제 배송 완료되지 않은 주문은 제외 (NaN 값 처리)
merged_df_cleaned = merged_df.dropna(subset=['order_delivered_customer_date']).copy()
merged_df_cleaned['actual_delivery_time'] = (merged_df_cleaned['order_delivered_customer_date'] - merged_df_cleaned['order_purchase_timestamp']).dt.days

# '예상 배송 완료일' 컬럼을 datetime 형식으로 변환 (orders_dataset에서 가져와 병합 필요)
# merged_df_cleaned에 orders_dataset의 'order_estimated_delivery_date' 컬럼 병합
merged_df_cleaned = pd.merge(merged_df_cleaned, orders_dataset[['order_id', 'order_estimated_delivery_date']], on='order_id', how='left')
merged_df_cleaned['order_estimated_delivery_date'] = pd.to_datetime(merged_df_cleaned['order_estimated_delivery_date'])

# '예상 배송 소요 시간' 계산
merged_df_cleaned['estimated_delivery_time'] = (merged_df_cleaned['order_estimated_delivery_date'] - merged_df_cleaned['order_purchase_timestamp']).dt.days

# 'estimated_delivery_time'이 음수인 경우 (예상일이 주문일보다 빠른 경우) 0으로 설정
merged_df_cleaned['estimated_delivery_time'] = merged_df_cleaned['estimated_delivery_time'].apply(lambda x: max(0, x))


# '예상 대비 지연 일수' 계산: 실제 배송 소요 시간 - 예상 배송 소요 시간
merged_df_cleaned['delay_against_estimated'] = merged_df_cleaned['actual_delivery_time'] - merged_df_cleaned['estimated_delivery_time']

# '배송 지연 여부' 컬럼 생성 (예상 대비 지연 일수가 0보다 큰 경우)
merged_df_cleaned['is_delayed'] = merged_df_cleaned['delay_against_estimated'] > 0

# 결과 확인
print("배송 지연 관련 컬럼 생성 결과 (처음 5개 행):")
display(merged_df_cleaned[['order_id', 'order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date', 'actual_delivery_time', 'estimated_delivery_time', 'delay_against_estimated', 'is_delayed']].head())

print("\n배송 지연 관련 컬럼 기술 통계량:")
display(merged_df_cleaned[['actual_delivery_time', 'estimated_delivery_time', 'delay_against_estimated']].describe())

print("\n배송 지연된 주문 비율:")
display(merged_df_cleaned['is_delayed'].value_counts(normalize=True) * 100)
```

```python
def model_impact_of_delay_reduction(percentage_reduction, merged_df, avg_review_by_delay):
    """
    Models the potential impact of a hypothetical percentage reduction in delivery delay
    on the average review score.

    Args:
        percentage_reduction: The hypothetical percentage reduction in delivery delay (e.g., 10 for 10%).
        merged_df: The DataFrame containing order and review information with 'delivery_delay_days'.
        avg_review_by_delay: DataFrame with average review score per delivery delay day.

    Returns:
        The projected overall average review score after the delay reduction.
    """
    # 1. Calculate the projected delivery delay for each order
    # Ensure the projected delay is not negative
    merged_df['projected_delay_days'] = merged_df['delivery_delay_days'] * (1 - percentage_reduction / 100)
    merged_df['projected_delay_days'] = merged_df['projected_delay_days'].apply(lambda x: max(0, x))

    # 2. Estimate the corresponding average review score based on the projected delay
    # Use the pre-calculated relationship (interpolate for non-exact matches)
    # Handle potential NaN values if projected_delay_days is outside the range of avg_review_by_delay
    estimated_review_scores = np.interp(
        merged_df['projected_delay_days'],
        avg_review_by_delay['delivery_delay_days'],
        avg_review_by_delay['review_score']
    )

    # 3. Return the average of these estimated review scores
    return np.mean(estimated_review_scores)

# Call the function with different hypothetical delay reduction percentages
reduction_percentages = [10, 20, 30]
print("잠재적 배송 지연 단축 효과 모델링:")
for percentage in reduction_percentages:
    projected_avg_review = model_impact_of_delay_reduction(percentage, merged_df.copy(), avg_review_by_delay) # Use a copy to avoid modifying the original df
    print(f"- 배송 시간 {percentage}% 단축 시 예상 평균 리뷰 점수: {projected_avg_review:.2f}")
```

**Reasoning**:
Segment customers based on their average delivery delay and average review score to identify those sensitive to delays. A simple approach is to categorize based on thresholds.



```python
# 배송 지연 민감도에 따라 고객 세그먼트화
# 임계값 정의 (예: 평균 배송 지연 7일 이상, 평균 리뷰 점수 3점 이하)
delay_threshold = 7
review_score_threshold = 3

def categorize_sensitivity(row):
    if row['average_delivery_delay_days'] > delay_threshold and row['average_review_score'] <= review_score_threshold:
        return 'High Sensitivity (Delay & Low Review)'
    elif row['average_delivery_delay_days'] > delay_threshold:
        return 'Medium Sensitivity (Delay Only)'
    elif row['average_review_score'] <= review_score_threshold:
        return 'Medium Sensitivity (Low Review Only)'
    else:
        return 'Low Sensitivity'

customer_metrics['sensitivity_segment'] = customer_metrics.apply(categorize_sensitivity, axis=1)

print("\n고객 세그먼트 분포:")
display(customer_metrics['sensitivity_segment'].value_counts())

# 각 고객 세그먼트의 특징 분석 (예: 평균 주문 횟수)
# merged_orders 데이터프레임 (이전에 생성된 고객별 주문 횟수 포함) 사용
segment_order_counts = pd.merge(customer_metrics, merged_orders[['customer_unique_id', 'order_count']].drop_duplicates(), on='customer_unique_id', how='left')

avg_order_count_by_segment = segment_order_counts.groupby('sensitivity_segment')['order_count'].mean().reset_index()
avg_order_count_by_segment.columns = ['sensitivity_segment', 'average_order_count']

print("\n세그먼트별 평균 주문 횟수:")
display(avg_order_count_by_segment)

# 세그먼트별 평균 배송 지연 일수 및 평균 리뷰 점수 확인
avg_metrics_by_segment = customer_metrics.groupby('sensitivity_segment').agg(
    mean_delivery_delay=('average_delivery_delay_days', 'mean'),
    mean_review_score=('average_review_score', 'mean')
).reset_index()

print("\n세그먼트별 평균 배송 지연 일수 및 평균 리뷰 점수:")
display(avg_metrics_by_segment)
```

```python
def model_impact_of_delay_reduction(percentage_reduction, merged_df, avg_review_by_delay):
    """
    Models the potential impact of a hypothetical percentage reduction in delivery delay
    on the average review score.

    Args:
        percentage_reduction: The hypothetical percentage reduction in delivery delay (e.g., 10 for 10%).
        merged_df: The DataFrame containing order and review information with 'delivery_delay_days'.
        avg_review_by_delay: DataFrame with average review score per delivery delay day.

    Returns:
        The projected overall average review score after the delay reduction.
    """
    # 1. Calculate the projected delivery delay for each order
    # Ensure the projected delay is not negative
    merged_df['projected_delay_days'] = merged_df['delivery_delay_days'] * (1 - percentage_reduction / 100)
    merged_df['projected_delay_days'] = merged_df['projected_delay_days'].apply(lambda x: max(0, x))

    # 2. Estimate the corresponding average review score based on the projected delay
    # Use the pre-calculated relationship (interpolate for non-exact matches)
    # Handle potential NaN values if projected_delay_days is outside the range of avg_review_by_delay
    estimated_review_scores = np.interp(
        merged_df['projected_delay_days'],
        avg_review_by_delay['delivery_delay_days'],
        avg_review_by_delay['review_score']
    )

    # 3. Return the average of these estimated review scores
    return np.mean(estimated_review_scores)

# Call the function with different hypothetical delay reduction percentages
reduction_percentages = [10, 20, 30]
print("잠재적 배송 지연 단축 효과 모델링:")
for percentage in reduction_percentages:
    projected_avg_review = model_impact_of_delay_reduction(percentage, merged_df.copy(), avg_review_by_delay) # Use a copy to avoid modifying the original df
    print(f"- 배송 시간 {percentage}% 단축 시 예상 평균 리뷰 점수: {projected_avg_review:.2f}")
```

**Reasoning**:
Segment customers based on their average delivery delay and average review score to identify those sensitive to delays. A simple approach is to categorize based on thresholds.

```python
# Calculate customer order counts and repeat customer flag if not already done
customer_order_counts = merged_df_full.groupby('customer_unique_id')['order_id'].nunique().reset_index()
customer_order_counts.columns = ['customer_unique_id', 'order_count']
customer_order_counts['is_repeat_customer'] = customer_order_counts['order_count'] > 1

# merged_df_full has order_id and customer_unique_id
merged_orders = merged_df_full[['order_id', 'customer_unique_id', 'order_purchase_timestamp', 'delivery_delay_days']].drop_duplicates(subset=['order_id', 'customer_unique_id'])
merged_orders = pd.merge(merged_orders, customer_order_counts, on='customer_unique_id', how='left')


# 배송 지연 민감도에 따라 고객 세그먼트화
# 임계값 정의 (예: 평균 배송 지연 7일 이상, 평균 리뷰 점수 3점 이하)
delay_threshold = 7
review_score_threshold = 3

def categorize_sensitivity(row):
    if row['average_delivery_delay_days'] > delay_threshold and row['average_review_score'] <= review_score_threshold:
        return 'High Sensitivity (Delay & Low Review)'
    elif row['average_delivery_delay_days'] > delay_threshold:
        return 'Medium Sensitivity (Delay Only)'
    elif row['average_review_score'] <= review_score_threshold: # Corrected typo
        return 'Medium Sensitivity (Low Review Only)'
    else:
        return 'Low Sensitivity'

customer_metrics['sensitivity_segment'] = customer_metrics.apply(categorize_sensitivity, axis=1)

print("\n고객 세그먼트 분포:")
display(customer_metrics['sensitivity_segment'].value_counts())

# 각 고객 세그먼트의 특징 분석 (예: 평균 주문 횟수)
# merged_orders 데이터프레임 (이전에 생성된 고객별 주문 횟수 포함) 사용
segment_order_counts = pd.merge(customer_metrics, merged_orders[['customer_unique_id', 'order_count']].drop_duplicates(), on='customer_unique_id', how='left')

avg_order_count_by_segment = segment_order_counts.groupby('sensitivity_segment')['order_count'].mean().reset_index()
avg_order_count_by_segment.columns = ['sensitivity_segment', 'average_order_count']

print("\n세그먼트별 평균 주문 횟수:")
display(avg_order_count_by_segment)

# 세그먼트별 평균 배송 지연 일수 및 평균 리뷰 점수 확인
avg_metrics_by_segment = customer_metrics.groupby('sensitivity_segment').agg(
    mean_delivery_delay=('average_delivery_delay_days', 'mean'),
    mean_review_score=('average_review_score', 'mean')
).reset_index()

print("\n세그멘트별 평균 배송 지연 일수 및 평균 리뷰 점수:")
display(avg_metrics_by_segment)
```

### 리뷰 키워드 분석 (텍스트 마이닝)

**Subtask:**
리뷰 텍스트 데이터(`review_comment_message`)를 사용하여 주요 키워드를 추출하고 시각화합니다. 리뷰 점수별(예: 긍정/부정 리뷰)로 나누어 키워드를 분석하면 더 유용한 인사이트를 얻을 수 있습니다.

**Reasoning:**
Perform text preprocessing on the review comments, including converting to lowercase, removing punctuation and stopwords, and then use a CountVectorizer to get the most frequent words. Visualize the top words for overall reviews.

```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
import string

# Download stopwords if not already downloaded
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')

# 텍스트 전처리 함수
def preprocess_text(text):
    if isinstance(text, str):
        # 소문자 변환
        text = text.lower()
        # 구두점 제거
        text = text.translate(str.maketrans('', '', string.punctuation))
        # 불용어 제거
        stop_words = set(stopwords.words('english'))
        text = ' '.join(word for word in text.split() if word not in stop_words)
        return text
    return '' # Handle non-string input


# 리뷰 메시지에 전처리 함수 적용
merged_reviews_df['cleaned_review_message'] = merged_reviews_df['review_comment_message'].apply(preprocess_text)

# 전체 리뷰에서 자주 등장하는 단어 추출
# CountVectorizer를 사용하여 단어 빈도 계산
vectorizer = CountVectorizer(max_features=50) # 상위 50개 단어 추출
X = vectorizer.fit_transform(merged_reviews_df['cleaned_review_message'])

# 단어 빈도를 데이터프레임으로 변환
word_counts = pd.DataFrame({'word': vectorizer.get_feature_names_out(), 'count': X.sum(axis=0).tolist()[0]})
word_counts = word_counts.sort_values(by='count', ascending=False)

print("전체 리뷰에서 자주 등장하는 단어 (상위 20개):")
display(word_counts.head(20))

# 자주 등장하는 단어 시각화
plt.figure(figsize=(12, 7))
sns.barplot(data=word_counts.head(20), x='count', y='word', palette='viridis')
plt.title('전체 리뷰에서 자주 등장하는 단어 (상위 20개)')
plt.xlabel('빈도 수')
plt.ylabel('단어')
plt.tight_layout()
plt.show()
```

### 리뷰와 매출 간의 상관관계 분석

**Subtask:**
리뷰 점수와 주문 결제 금액(`payment_value`) 간의 관계를 분석하고 시각화하여 고객 만족도가 매출에 미치는 영향을 파악합니다. 이를 통해 리뷰 점수 개선의 중요성을 강조할 수 있습니다.

**Reasoning:**
Merge the `merged_reviews_df` with `order_payments` to include payment value information. Then, group by review score and calculate the average payment value for each score. Finally, visualize the relationship between average payment value and review score using a bar plot.

```python
# merged_reviews_df와 order_payments 데이터프레임을 'order_id' 컬럼을 기준으로 inner join으로 병합
# 각 order_id에 대해 여러 payment가 있을 수 있으므로, payment_value의 합계를 사용하거나 order_id 기준으로 unique한 payment를 선택해야 함.
# 여기서는 단순화를 위해 order_id 기준으로 order_payments를 그룹화하여 payment_value의 합계를 사용합니다.
order_payments_agg = order_payments.groupby('order_id')['payment_value'].sum().reset_index()
order_payments_agg.columns = ['order_id', 'total_payment_value']

merged_reviews_with_payment = pd.merge(merged_reviews_df, order_payments_agg, on='order_id', how='inner')

# 리뷰 점수별 평균 결제 금액 계산
avg_payment_by_review_score = merged_reviews_with_payment.groupby('review_score')['total_payment_value'].mean().reset_index()
avg_payment_by_review_score.columns = ['review_score', 'average_payment_value']

print("리뷰 점수별 평균 결제 금액:")
display(avg_payment_by_review_score)

# 리뷰 점수별 평균 결제 금액 시각화
plt.figure(figsize=(8, 5))
sns.barplot(data=avg_payment_by_review_score, x='review_score', y='average_payment_value', palette='viridis')
plt.title('리뷰 점수별 평균 결제 금액')
plt.xlabel('리뷰 점수')
plt.ylabel('평균 결제 금액')
plt.show()

# 리뷰 점수와 결제 금액 간의 상관관계 분석 (산점도 및 상관계수)
plt.figure(figsize=(8, 5))
sns.scatterplot(data=merged_reviews_with_payment, x='review_score', y='total_payment_value', alpha=0.5)
plt.title('리뷰 점수와 총 결제 금액의 관계')
plt.xlabel('리뷰 점수')
plt.ylabel('총 결제 금액')
plt.show()

# 상관계수 계산
correlation = merged_reviews_with_payment['review_score'].corr(merged_reviews_with_payment['total_payment_value'])
print(f"\n리뷰 점수와 총 결제 금액 간의 상관계수: {correlation:.2f}")
```

```python
# Re-merge dataframes to include all necessary columns for analysis
# Start with orders_dataset and order_reviews
merged_df_full = pd.merge(orders_dataset, order_reviews, on='order_id', how='inner')

# Merge with customers
merged_df_full = pd.merge(merged_df_full, customers, on='customer_id', how='inner')

# Merge with order_items to get product_id and price
merged_df_full = pd.merge(merged_df_full, order_items, on='order_id', how='inner')

# Select necessary columns including 'product_id' and 'seller_id'
merged_df_full = merged_df_full[['order_id', 'customer_id', 'order_purchase_timestamp', 'order_delivered_customer_date', 'review_score', 'customer_unique_id', 'product_id', 'seller_id', 'price']]

# Convert date columns to datetime format
merged_df_full['order_purchase_timestamp'] = pd.to_datetime(merged_df_full['order_purchase_timestamp'])
merged_df_full['order_delivered_customer_date'] = pd.to_datetime(merged_df_full['order_delivered_customer_date'])

# Calculate delivery delay in days
merged_df_full['delivery_delay'] = merged_df_full['order_delivered_customer_date'] - merged_df_full['order_purchase_timestamp']
merged_df_full['delivery_delay_days'] = merged_df_full['delivery_delay'].dt.days
merged_df_full['delivery_delay_days'] = merged_df_full['delivery_delay_days'].apply(lambda x: max(0, x) if pd.notna(x) else np.nan)
```

```python
# 고객별 평균 리뷰 점수와 평균 배송 지연 일수 계산
customer_metrics = merged_df_full.groupby('customer_unique_id').agg(
    average_review_score=('review_score', 'mean'),
    average_delivery_delay_days=('delivery_delay_days', 'mean')
).reset_index()

print("고객별 평균 리뷰 점수 및 평균 배송 지연 일수 (상위 10개 고객):")
display(customer_metrics.head(10))
```

```python
# Calculate customer order counts and repeat customer flag if not already done
customer_order_counts = merged_df_full.groupby('customer_unique_id')['order_id'].nunique().reset_index()
customer_order_counts.columns = ['customer_unique_id', 'order_count']
customer_order_counts['is_repeat_customer'] = customer_order_counts['order_count'] > 1

# merged_df_full has order_id and customer_unique_id
merged_orders = merged_df_full[['order_id', 'customer_unique_id', 'order_purchase_timestamp', 'delivery_delay_days']].drop_duplicates(subset=['order_id', 'customer_unique_id'])
merged_orders = pd.merge(merged_orders, customer_order_counts, on='customer_unique_id', how='left')


# 배송 지연 민감도에 따라 고객 세그먼트화
# 임계값 정의 (예: 평균 배송 지연 7일 이상, 평균 리뷰 점수 3점 이하)
delay_threshold = 7
review_score_threshold = 3

def categorize_sensitivity(row):
    if row['average_delivery_delay_days'] > delay_threshold and row['average_review_score'] <= review_score_threshold:
        return 'High Sensitivity (Delay & Low Review)'
    elif row['average_delivery_delay_days'] > delay_threshold:
        return 'Medium Sensitivity (Delay Only)'
    elif row['average_review_score'] <= review_score_threshold:
        return 'Medium Sensitivity (Low Review Only)'
    else:
        return 'Low Sensitivity'

customer_metrics['sensitivity_segment'] = customer_metrics.apply(categorize_sensitivity, axis=1)

print("\n고객 세그먼트 분포:")
display(customer_metrics['sensitivity_segment'].value_counts())

# 각 고객 세그먼트의 특징 분석 (예: 평균 주문 횟수)
# merged_orders 데이터프레임 (이전에 생성된 고객별 주문 횟수 포함) 사용
segment_order_counts = pd.merge(customer_metrics, merged_orders[['customer_unique_id', 'order_count']].drop_duplicates(), on='customer_unique_id', how='left')

avg_order_count_by_segment = segment_order_counts.groupby('sensitivity_segment')['order_count'].mean().reset_index()
avg_order_count_by_segment.columns = ['sensitivity_segment', 'average_order_count']

print("\n세그먼트별 평균 주문 횟수:")
display(avg_order_count_by_segment)

# 세그먼트별 평균 배송 지연 일수 및 평균 리뷰 점수 확인
avg_metrics_by_segment = customer_metrics.groupby('sensitivity_segment').agg(
    mean_delivery_delay=('average_delivery_delay_days', 'mean'),
    mean_review_score=('average_review_score', 'mean')
).reset_index()

print("\n세그멘트별 평균 배송 지연 일수 및 평균 리뷰 점수:")
display(avg_metrics_by_segment)
```
